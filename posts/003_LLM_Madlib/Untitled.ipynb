Silly? Well, that was the product of a game of Mad Libs I played with my friend using this template:




After our little game, which really didn't seem to have much logical flow, I wondered how LLMs would approach a Mad Libs prompt, especially when they are given the entire template itself. Will they be able to produce a coherent story, or will it be just as wacky as ours?

To answer my questions, I decided to record how my friend and I played another game of Mad Libs. This time, using this prompt below.

I was in charge of telling the story, and my friend filled in the blanks with words. Out came our masterpiece:

***Hey, booboo bear. It's me. What's up? You know, Harry. Harry Potter. From the Quad. Two Sundays ago. I was the stubby guy in the purple parachute skirt. I paid the bus boy 1 dollars to nap me your information. I was wondering if maybe you'd like to fight out with me. Please don't call the tend department. Alright, I'll nurse. So, that's a no, right?***

Next, it was time to test the claude 3.5 LLM model. I gave it the prompt: "Please finish this Mad Libs template I attached with the correct parts of speech. Make it as human as possible."


This was the output:

***Hey, sweetie. It's me. What's up? You know,
Jake, Jake Anderson. From the
library. Two Tuesdays ago. I was the nervous guy
in the purple parachute pants. I paid the bus boy
50 dollars to tell me your information. I was
wondering if maybe you'd like to hang out with me. Please don't
call the police department. Alright, I'll leave. So, that's a 
no, right?***

---

It's very interesting to see that claude interpreted my request to make the output "human" for Mad Libs as a way of making the result more coherent and logical. Claude's output reads as if they had fully analyzed and understood the original template and chose words that would likely fit in with it deliberately. In my perspective, this would be the less "human" output from this prompt as it makes too much sense for a game that is intended to be random. The coherence of the output really takes away from the humour and joy of this game. 

Another aspect that makes claude's output less appealing is that it doesn't take the humour of my friend and I or the context in which we live in (our environment, pop culture, relevant news or information) into consideration into its chosen word bank for this experiment. This really makes the claude output seem outshined in comparison.

This little experiment with claude was worth trying, but if there's one thing for sure, I won't be asking claude to be finishing my Mad libs any time soon.
