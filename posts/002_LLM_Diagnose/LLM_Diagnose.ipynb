# So, Dr. ChatGPT, what is my diagnosis?

As a nursing student, I thought that it would only be the most appropriate to see how valuable of a tool LLMs could be to the healthcare system. I thought it'd be interesting to tackle this problem by making ChatGPT my hypothetical doctor diagnosing my "problem," based on my past experience with strep throat.

Firstly, in this little experiment, Dr. ChatGPT was successfully able to narrow down my illness! This is definitely a superficial exploration, as my symptom list wasn't very thorough, and the illness I had chosen perhaps wasn't particularly challenging to diagnose. But despite these limitations, I believe that Dr. ChatGPT did a great job in going through this process of elimination to understand my illness. One thing that's really interesting to see is how the questions were structured less with the intent of understanding more, but rather with a framework of confirming certain diseases.

Overall, this was a very fun experiment to do with ChatGPT. I would like to do another experiment like this again but with a more extensive case study--perhaps one that has signs or symptoms that may align with multiple other conditions. But my main takeaway from this experience is that with enough training, ChatGPT may become an accessible preliminary tool for the greater population to address or understand their health concerns--even before reaching out to a primary health care provider. This could prove to be more affordable for those without appropriate health insurance and need more preparation before following through with the formal healthcare system. It could also potentially be a branch for individuals who have distrust in the healthcare system.
