We learn from an early age to be polite--saying the word “please” or “thank you,”--in hopes of receiving more kind responses. But what about when we’re talking to machines? It is curious to learn whether wording prompts politely makes any difference in the answers they get. Does saying “please” unlock better results, or is it just a habit carried over from human conversation? And well, there's only one way to find out--try it for ourselves.

# The Method

I decided to compare ChatGPT's reponse to asking it how to answers an explanatory question in two separate chats. I repeated this 3 separate times to see how the responses would vary. I chose a question using ChatGPT, asking it what it thought would be an appropriate question to ask for this little experiment, and I liked its results.

The two questions:

> Explain photosynthesis in a concise form.

> Please explain photosynthesis in a concise form.

# Results


It’s intriguing to note that one of the main differences between these two outputs is that, as I continued to ask ChatGPT the same question, the answers became progressively more brief and concise. This gives the impression that the model was refining its responses over time, almost as if it “remembered” having already provided an explanation and was adjusting to avoid redundancy. The shortening of responses could also suggest that the model is optimizing for efficiency once it has established the key elements of the answer, prioritizing clarity over repetition.

Overall, I infer that there is no difference in the quality or quantity of output given by ChatGPT when prompted politely or not. Both sets of responses maintain the same level of accuracy, clarity, and usefulness, with only slight variations in style or depth. This suggests that the inclusion of the word “please” does not directly influence the factual content or thoroughness of the explanation, but rather the model’s internal balancing of conciseness and variety (temperature) in repeated answers.
